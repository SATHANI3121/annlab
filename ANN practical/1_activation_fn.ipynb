```python
import numpy as np
import matplotlib.pyplot as plt

def plot_activation_function(x, y, title):
    plt.plot(x, y)
    plt.xlabel('X value')
    plt.ylabel(f'{title}(x)')
    plt.title(f"{title} function")
    plt.grid(True)
    plt.show()

# Step function
x = np.linspace(-5, 5)
y_step = np.heaviside(x, 0.5)
plot_activation_function(x, y_step, 'Step')

# Sigmoid function
x = np.linspace(-10, 10)
y_sigmoid = 1 / (1 + np.exp(-x))
plot_activation_function(x, y_sigmoid, 'Sigmoid')

# ReLU function
x = np.linspace(-10, 10)
y_relu = np.maximum(0, x)
plot_activation_function(x, y_relu, 'ReLU')

# Leaky ReLU function
def leaky_relu(x):
    alpha = 0.1
    return np.where(x >= 0, x, alpha * x)

x = np.linspace(-10, 10)
y_leaky_relu = leaky_relu(x)
plot_activation_function(x, y_leaky_relu, 'Leaky ReLU')

# Tanh function
x = np.linspace(-10, 10)
y_tanh = np.tanh(x)
plot_activation_function(x, y_tanh, 'Tanh')

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

x = np.linspace(-10, 10)
y_softmax = softmax(x)
plot_activation_function(x, y_softmax, 'Softmax')

# Binary step function
x = np.linspace(-10, 10)
y_binary_step = np.heaviside(x, 0.5)
plot_activation_function(x, y_binary_step, 'Binary Step')
```
